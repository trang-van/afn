### Clean and Match Oakland Files (No PPP Loan Data)

```python
import pandas as pd
import numpy as np
import re
import warnings
import os
from os.path import dirname, abspath

warnings.filterwarnings('ignore')
```

```python
from cleanco import cleanco
```

```python
def merge_df(df1, df2, on_col1, on_col2):
    return df1.merge(df2, how='left', left_on=on_col1, right_on=on_col2)
```

```python
d = dirname(dirname(abspath(os.getcwd())))
```

```python
oakland_file = d + '/data-to-match/Oakland/Business_database_Oakland.csv'
oakland_active_bus_file = d + '/data-to-match/Oakland/Oakland_ActiveBusinessAccounts_to_match.csv'
```

```python
oakland_df = pd.read_csv(oakland_file, encoding='mac_roman')
oakland_df.head()
```

```python
header_row = 2
oakland_df.columns = oakland_df.iloc[header_row]
oakland_df.drop([0,1,2], inplace=True)
oakland_df.dropna(axis='columns', how='all', inplace=True)
oakland_df.dropna(axis='rows', how='all', inplace=True)
```

```python
oakland_df.head()
```

```python
# AFter Dropping NaN Columns
oakland_df.columns
```

```python
oakland_df
```

```python
oakland_df['Company Name - Clean'] = oakland_df['Company name Latin alphabet'].apply(str).apply(lambda x: cleanco(x).clean_name())
```

```python
oakland_df.head()
```

```python
active_bus_df = pd.read_csv(oakland_active_bus_file, encoding='mac_roman')
active_bus_df.head()
```

```python
active_bus_df['Company Name - Clean'] = active_bus_df['Company name Latin alphabet'].apply(str).apply(lambda x: cleanco(x).clean_name())
```
