```python
import pandas as pd
import numpy as np
```

```python
CPS = pd.read_csv('cps_00006.csv.gz')
```

# Categories

## Race

```python
CPS['RACE'].unique()
```

```python
#Create new column called race category. Create conditions for categories and assign values to each condition
conditions = [
    (CPS['RACE'] == 100),(CPS['RACE'] == 200),(CPS['RACE'] == 300),
    (CPS['RACE'] == 651),(CPS['RACE'] == 652),(CPS['RACE'] >= 801)
    ]
values = ['White', 'Black', 'American Indian/Native','Asian', 'Hawaiian/Pacific Islander', 'Two or more races']

CPS['RACE_CAT'] = np.select(conditions, values)

CPS.head(10)
```

```python
CPS['RACE_CAT'].unique()
```

```python
CPS['HISPAN'].unique()
```

```python
#Create new column called hispanic category. Create conditions for categories and assign values to each condition
conditions = [
    (CPS['HISPAN'] == 0),(CPS['HISPAN'] >= 100)
    ]
values = ['Not Hispanic/ Latinx', 'Hispanic/ Latinx']

CPS['HISPAN_CAT'] = np.select(conditions, values)

CPS.head(10)
```

```python
CPS['HISPAN_CAT'].unique()
```

```python
#Create new column called Asian category. Create conditions for categories and assign values to each condition
conditions = [
    (CPS['ASIAN'] == 99.),(CPS['ASIAN'] == 30.),(CPS['ASIAN'] == 10.),
    (CPS['ASIAN'] == 70.),(CPS['ASIAN'] == 40.),(CPS['ASIAN'] == 20.),
    (CPS['ASIAN'] == 50.),(CPS['ASIAN'] == 60.),
    ]
values = ['NIU', 'Filipino', 'Asian Indian','Other Asian', 'Japanese', 'Chinese', 'Korean', 'Vietnamese']

CPS['ASIAN_CAT'] = np.select(conditions, values)

CPS.head(10)
```

## Employment

```python
#Create new column called Self Employed category. Create conditions for categories and assign values to each condition
conditions = [
    (CPS['CLASSWKR'] == 10.),(CPS['CLASSWKR'] == 13.),(CPS['CLASSWKR'] == 14.),
    (CPS['CLASSWKR'] == 20.),(CPS['CLASSWKR'] == 21.),(CPS['CLASSWKR'] == 22.),
    (CPS['CLASSWKR'] == 23.),(CPS['CLASSWKR'] == 24.), (CPS['CLASSWKR'] == 25.),
    (CPS['CLASSWKR'] == 26.), (CPS['CLASSWKR'] == 27.), (CPS['CLASSWKR'] == 28.), 
    (CPS['CLASSWKR'] == 29.), (CPS['CLASSWKR'] == 00.), (CPS['CLASSWKR'] == 99.),
    ]
values = ['Self_Employed', 'Self_Employed', 'Self_Employed', 
          'Not_Self_Employed', 'Not_Self_Employed', 'Not_Self_Employed', 
          'Not_Self_Employed', 'Not_Self_Employed', 'Not_Self_Employed', 
          'Not_Self_Employed', 'Not_Self_Employed', 'Not_Self_Employed', 
          'Not_Self_Employed', 'NA', 'NA']

CPS['SELFEMPLOYED_CAT'] = np.select(conditions, values)

CPS.head(10)
```

# Employment by Industry

## 2020 - US

```python
#Extract year 2020 

CPS_2020 = CPS[(CPS['YEAR']==2020)]
CPS_2020.tail(5)
```

```python
CPS_2020_ASECFLAG = CPS_2020[(CPS_2020['ASECFLAG']==1.0)]
CPS_2020_ASECFLAG.head(5)
```

```python
#Employment counts 2020

pd.crosstab(index=CPS_2020_ASECFLAG['SELFEMPLOYED_CAT'], columns='count', margins=True, margins_name="Total")
```

```python
CPS_2020_ASECFLAG.groupby(['IND_CAT', 'SELFEMPLOYED_CAT']).agg({'SELFEMPLOYED_CAT' : 'count'})
```

```python
Ind_Emp_group = CPS_2020_ASECFLAG.groupby(["IND_CAT","SELFEMPLOYED_CAT"]).agg({"SELFEMPLOYED_CAT" : "count"})
```

```python
# proportion within industry

Ind_Emp_group.groupby(level=[0]).apply(lambda x: 100*x/x.sum())
```

```python
# percent all industries (ex: .03% of self employed ppl work in the Bookstore industry)

Ind_Emp_group.groupby(level=[1]).apply(lambda x: 100*x/x.sum())
```

## 2020 - Race Categories - US Level

```python
# Race counts 2020

pd.crosstab(index=CPS_2020_ASECFLAG['RACE2_CAT'], columns='count', margins=True, margins_name="Total")
```

```python
# 2020 Number self employed vs race cat

pd.crosstab(index=CPS_2020_ASECFLAG['SELFEMPLOYED_CAT'], columns=CPS_2020_ASECFLAG['RACE2_CAT'], margins=True, margins_name="Total")
```

```python
# create df ind+emp+Race

Ind_Emp_Race_group = CPS_2020_ASECFLAG.groupby(["IND_CAT", 'SELFEMPLOYED_CAT', "RACE2_CAT"]).agg({"SELFEMPLOYED_CAT" : "count"})
```

```python
# employment count by race info

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(Ind_Emp_Race_group)
```

```python
# Percentage all industries (ex: .0128% of self employed ppl in Bookstore industry identify as Asian)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.5f}'.format
display(Ind_Emp_Race_group.groupby(level=[1]).apply(lambda x: 100*x/x.sum()))
```
